# Image scraper for JapanLA

This script is an image scraper for the online shop Japan LA to collect data of authentic plushies.

Import all necessary models to make the image scraper work and save all the contents

```python
import requests
from bs4 import BeautifulSoup
import os
from builtins import open, bytes
import uuid
import time
import random
import string
```

The scraper is into a function so it can be executed multiple times by just referencing it. The function accept a URL with the parameter 'url' <br>
1. A response to the URL is made, to which the page HTML source is parsed in test
2. The script finds every image element in the HTML that include a source, and gets the URLs of these
3. A folder where the images go is refenced in a variable. If the folder does not exist, it will be created first
4. Using a 'for loop', the images are downloaded from their source URLs and saved in the aforementioned folder. The images are retrieved and saved by creating the file first and copying the source data from the images in it as binary data, which is used for non-text files
<br> <br> 
The images are named according to the authenticity. To prevent the images from overridding eachother, a randomized string of characters is added at the end of a filename. The script also checks if the source URL starts with 'http' to download the images. If not the case, the protocol indicator is added at the start of the URL.

```python
def download_images(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    
    protocol_indicator = "https:"

    imgs = soup.find_all("img", {"src": True})

    folder_name = 'test'
    if not os.path.exists(folder_name):
            os.makedirs(folder_name)

    for i, img in enumerate (imgs):
            src = img["src"]
            
            random_string = ''.join(random.choice(string.ascii_lowercase) for i in range(5))
            filename = random_string
            
            if not src.startswith("http"):
                src = protocol_indicator + src
            response = requests.get(src)

            with open(f"{folder_name}/authentic_{filename}.jpg", 'wb') as f:
                f.write(response.content)
```

The function is called with the URL in the parameter, to which all the images are downloaded in the folder. As there are multiple pages of listings on the site, the function would have to be called for each one

```python
download_images("https://www.japanla.com/collections/plush")
download_images("https://www.japanla.com/collections/plush?page=2")
download_images("https://www.japanla.com/collections/plush?page=3")
download_images("https://www.japanla.com/collections/plush?page=4")
download_images("https://www.japanla.com/collections/plush?page=5")
download_images("https://www.japanla.com/collections/plush?page=6")
download_images("https://www.japanla.com/collections/plush?page=7")
download_images("https://www.japanla.com/collections/plush?page=8")
download_images("https://www.japanla.com/collections/plush?page=9")
download_images("https://www.japanla.com/collections/plush?page=10")
download_images("https://www.japanla.com/collections/plush?page=11")
```
