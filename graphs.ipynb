# Python graphs of performance progression of bootleg/ authentic identifier

To keep track of my model's performance over the course of the semester, I created Excel sheets with data about the validation/training loss and accuracy of the model. I'll then graph the data in line graphs to easily visualize the progress and compare the changes made in the model with eachother.
MatPlotLib and Pandas are imported to be able to create graphs from data

```python
import matplotlib.pyplot as plt
import pandas as pd
```

The Excel sheet containing the data

```python
df=pd.read_excel(r"C:\Users\jenny\Desktop\minor\finetuning results determining real or bootleg.xlsx")
```

Print the data in the log out

```python
print(df)
```

A graph about the model's accuracy data is plotted in a line graph. A title, legend, and labels are added to make the graphs easier to read. To not cause confusion at first glance, the y coordinate begins at 0

Overall, the accuracy remains relatively high in the beginning of the project, with 93% being the highest peak and 71% the lowest which is a 22% difference. 

In the beginning, I worked with a data set that contained unedited images that contained the background and other objects as well as images that weren't even plushies. at the 5th epoch, I removed the background of the images in the data set which caused the accuracy to decrease since the model no longer takes the background in account. While the measurement of the accuracy has decreased, the calculations of the images are now more accurate.

At the 13th epoch however, the accuracy decreases significally, with 59% being the lowest. This was the time when i reconstructed my data set and divided in folders for determining the character and then the authenticity of a plush. Since the folder for authenticity is also divided in characters and is much smaller than the old data set, I suspect that the change in accuracy is due to that. But like the 5th epoch, this also means that the model can predict more accurately since it now compared similiar items instead of numerous different items in the same folder.

```python
plt.plot(df['epoch'],df['accuracy'])

plt.title('Accuracy of model in %')

plt.xlabel('Epochs')
plt.ylabel('Accuracy in %')

line_1, = plt.plot(df['epoch'],df['accuracy'], label='Accuracy', color='hotpink')
plt.legend(handles=[line_1])

plt.ylim(ymin=0,ymax=1)
plt.grid()
plt.show()
```

A graph about the model's validation and training loss data is plotted in a line graph. A title, legend, and labels are added to make the graphs easier to read. To not cause confusion at first glance, the y coordinate begins at 0

The loss of both types are first in roughly within the same range of each other. What's peculiar is that the training loss peaked twice above the validation loss, before the latter outnumbers the training loss  by far at the end. The peak in training loss could be due to the model being 'overtrained', basically memorizing instead of detecting patterns. <br>
At the end of the graph, the differences between the losses increases even further. At that point, I had divided the data set into a characters consisting of a very small set of images. I suspect because of that, the model has become 'undertrained' and thus gives incorrect predictions alot

```python
plt.plot(df['epoch'],df['valid_loss'])
plt.plot(df['epoch'],df['train_loss'])

plt.xlabel('Epochs')
plt.ylabel('Loss')

plt.title('Validation and Training loss in model')

line_1, = plt.plot(df['epoch'],df['valid_loss'], label='Validation loss', color='red')
line_2, = plt.plot(df['epoch'],df['train_loss'], label='Training loss',color= 'blue')
plt.legend(handles=[line_1,line_2])

plt.grid()
plt.show()
```
